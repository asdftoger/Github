```python
# -*- coding: utf-8 -*-
"""
Created on Mon Feb 06 20:38:11 2017

@author: John
"""
import numpy as np
import numpy.linalg as lin
import matplotlib.pyplot as plt
from scipy import misc

#Loading the image, flatten converts the image to B&W
img = misc.imread('Dog.png',flatten = True)


#SVD on image
U,S,V = lin.svd(img)
#reshaping sizes
pic_dim = U.shape[0],V.shape[0]

#X shows the cumulative weight of singular values in S
X = np.cumsum(S)
X = X/sum(S)
plt.plot(X)

plt.xlabel('Singular value,number')
plt.ylabel('Normalized weight')

plt.title('Cumalative Weight of singular values')
plt.show()
```

```python

rank = [i for i in np.arange(1,np.sqrt(len(S)))] #changes rank of matrices to compare compression

for num,i in enumerate(rank):
    #s_mod creates a copy of S to work with
    s_mod = np.copy(S)
    #Extracts singular values
    s_mod[i:] = 0
    #Diagonalizes the matrix
    s_mod = np.diag(s_mod)
    #Resize to picture dimension and perform SVD
    s_mod.resize(pic_dim)
    A = np.dot(U,np.dot(s_mod,V))
    
    #Subplot plot, plotting 4x7 images
    plt.subplot(4,7,num+1)
    plt.text(0.5, 0.5, 'Rank:{}'.format(i), verticalalignment='bottom', horizontalalignment='left', fontsize=5)
    plt.axis('off')
    plt.imshow(A, interpolation = 'nearest',cmap = plt.cm.gray)
    
#Saves the plot to png
plt.savefig('SVD_on_dog.png',dpi = 1000)
plt.show()
```
