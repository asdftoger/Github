```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
```

```python
#Plot a 3x3 subplots containing polynomials

#Preparing the data, and number of subplots
x = np.linspace(-5,5)
n = [i+1 for i in range(9)]

```

#The money shot of the program

```python
#num - is the index number
#i - is the value
for num,i in enumerate(n):
    
    
    #Creates 3x3 grid, places subplot in each spot depending on num
    plt.subplot(3,3,num + 1)
    
    #Create text to annotate which degree polynomial
    plt.title('Degree {}'.format(i),size = 10)
    
    plt.axis('off')
    #Plot the i degree polynomial
    plt.plot(x,x**i)
    
plt.show()
```
