Setting up contour plot data, in this case a hemispherical plot

```python
R = 2
spacing = 100
x = np.linspace(-R,R,spacing)
y = np.linspace(-R,R,spacing)

X,Y = np.meshgrid(x,y)
Z = (R**2 - (X**2 + Y**2) )**0.5
```

```python
plt.figure()
_ = plt.contour(X,Y,Z, colors= 'k')
plt.show()
```

```python
#3D plot of the same data
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
_ = ax.plot_surface(X,Y,Z, color = 'r')
plt.show()
```
